## Azure MachineLearningの準備
まず、Azure MachineLearning（Azure ML）を使うための事前準備や、セットアップ作業について説明します。  
事前準備では「Microsoftアカウント」を取得し、次にAzure MLのFreeプランへの利用登録を行います。

既にAzure MLのアカウントをお持ちの場合は、本資料は飛ばしていただいて大丈夫です。

### Microsoftアカウントの取得
まず、「Microsoftアカウント」を取得します。  
既にアカウントをお持ちの場合は、**Azure MLのFreeプランへの利用登録**から進めてください。

MicrosoftのHP（https://azure.microsoft.com/ja-jp/free/） にアクセスし、【無料で始める】をクリックしましょう。


![](images/2_4_0_01.png)

図：Microsoftアカウントの登録画面にアクセス

【作成】をクリックします


![](images/2_4_0_02.png)

図：アカウントの作成画面

メールアドレスの入力画面に移りますので  
登録に使用したいメールアドレスを入力し、【次へ】をクリックします。


![](images/2_4_0_03.png)

図：メールアドレスの入力画面

パスワードを入力し、【次へ】をクリックします。


![](images/2_4_0_04.png)

図：パスワードの入力画面

登録したメールアドレス宛にセキュリティーコードが届きますので、  
【コードの入力】欄に入力し、【次へ】をクリックします。

新製品に関する情報などメール配信が欲しい方は、チェックボックスにチェックを入れておきましょう。


![](images/2_4_0_05.png)

図：セキュリティーコードの入力

表示されている文字（アルファベット・数字）を【表示されている文字を入力してください】の欄に入力しましょう。  
大文字/小文字も区別されるので、入力の際には注意が必要です。  

もし、文字の判読が難しい場合は【新規】をクリックすると、新しい文字が表示されます。

入力が終わりましたら、【次へ」をクリックします。


![](images/2_4_0_06.png)

図：機械でないことの確認

サインインした状態を維持するか、確認が出てきますので【はい】をクリックします。  
そうすることで、次回以降はパスワードを入力せずにログインが可能となります。


![](images/2_4_0_07.png)

図：サインイン維持の確認画面

これでMicrosoftアカウントの取得が完了しました。

### Azure MLのFreeプランへの利用登録

Azure MLの紹介ページ(https://azure.microsoft.com/ja-jp/services/machine-learning-studio/)にアクセスします。   
そして、【今すぐご利用ください】をクリックして先ほど作成したMicrosoftアカウントでサインインしましょう。


![](images/2_4_0_08.png)

図：Azure MLの紹介ページ

料金プランの選択画面が表示されるので中央のFreeプランの【Sign In】をクリックしましょう。


![](images/2_4_0_09.png)

図：料金プランの選択

以下の画面が表示されれば、Azure MLの利用登録は完了となります。


![](images/2_4_0_10.png)

図：Azure MLの利用画面

以上でAzure Machine Learningの準備が完了となります。
