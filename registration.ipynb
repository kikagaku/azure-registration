# Custom Visionの準備

進め方の手順としては以下の通りです。

- Microsoftアカウントの取得
- Azure従量課金にサインアップ

既にアカウンを取得できている方は、２点目の「Azure従量課金にサインアップ」から進めていきましょう。

## Microsoftアカウントの取得

まず、「Microsoftアカウント」を取得します。  
[Custom Vision](https://azure.microsoft.com/ja-jp/services/cognitive-services/custom-vision-service/)

「はじめる」をクリックします。

![](images/1-1-1_01.png)

「SIGN IN」をクリックします。

![](images/1-1-1_02.png)

アカウントをお持ちでない方は「作成」をクリックします。  
既にお持ちの方はサインインを進めてください。

![](images/1-1-1_03.png)

まずサインインするために「メールアドレス」または「電話番号」を入力しましょう。

![](images/1-1-1_04.png)

「パスワード」を決めましょう。

![](images/1-1-1_05.png)

確認メールが届くので、確認メールに記載の数字を入力しましょう。

![](images/1-1-1_06.png)

ロボットではないことを音声または文字で認証をしていきましょう。

![](images/1-1-1_07.png)

そのまま「はい」で進めていきましょう。

![](images/1-1-1_08.png)

「アプリの許可」を行いましょう。

![](images/1-1-1_09.png)

チェックボックスにチェックをつけて「I agree」をクリックしましょう。

![](images/1-1-1_10.png)

クレジットカードの登録を行うため「Sign up for Azure」をクリックしましょう。

![](images/1-1-1_11.png)


## Azure従量課金にサインアップ

まずはユーザの情報を入力していきましょう。

![](images/1-1-1_12.png)

続いて、本人確認を行います。「電話をする」をクリックするとアメリカ合衆国から電話かかってきて音声案内で数字を読み上げてくれます。「テキストメッセージを送信」の場合はメッセージが送られてきます。

![](images/1-1-1_13.png)

上記で確認した数字を「確認コード」に入力しましょう。

![](images/1-1-1_14.png)

クレジットカードの情報を入力しましょう。

![](images/1-1-1_15.png)

「テクニカルサポートがないか、...」にチェックをつけましょう。

![](images/1-1-1_16.png)

最後「同意する」ところにチェックをつけて「サインアップ」を行いましょう。

![](images/1-1-1_17.png)

Azureの登録が完了です。

![](images/1-1-1_18.png)

お疲れ様でした。
